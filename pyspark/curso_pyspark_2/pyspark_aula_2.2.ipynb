# Criando um ambiente local
* Vamos montar todo um ambiente local. Instalar o Spark e o MINIO na maquina.
* MINIO eh um servidor de objetos, que vamos utilizar para criar Data Lakes.
  * Landing Zone + Processing Zone + Curated Zoone

### Instalando o MinIO localmente
* Instalar atraves do site oficial do MinIO a partir do seu SO.
* Ao instalar o MinIO, podemos criar um servidor local passando `$ minio server "path"`
  * Abrira entao uma pagina no browser com o MinIO. No terminal teremos o endereco no API.
* Podemos criar as zonas (buckets) do nosso dl na pagina do MinIO do browser.
