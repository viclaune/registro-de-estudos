# O que é o Apache Spark?
* Projeto Open Source
* Engine de processamento em ambiente distribuido (cluster de computadores)
  * Spark se torna escalavel, podemos ter 1 ou centenas de nos (computadores)

* APIs do Spark
  * Permite trabalhar de diversas formas e se comunica com diversas ferramentas.
    * Kafka, Hadoop HDFS, Cassandra, PostgreSQL, elastic, DW, Cloud...
    * R, Python, Scala, R, SQL...

* Spark faz o processamento de dados em memória em ambiente distribuído. Também faz processamento em Streaming.

# O que é Databricks?
* Eh uma solução em Cloud que permite processar cluster de maquinas Spark.
* As grandes clouds levaram o databricks para dentro delas.

# Dataframes e Particoes no Spark
* Dataframes sao estruturas em memoria, como tabelas (linhas e colunas).
* Uma diferenca entre trabalhar com dataframes no pandas e no spark eh que no spark a tabela eh particionado pelos nos do cluster.
* Essa diferenca eh tanto para leitura quanto para escrita.

# Boas praticas para gestao de Data Lakes
* Data Lake Zones:
  * Landind Zone - Zona em que os dados chegam.
  * Processing - Dados que ja sofreram alguma tipo de processamentp
  * Refined - Dados preparados para serem utilizados.

# Apache Parquet
* Formato de dados que foi desenvolvido para trabalhar com Big Data.
* Armazena dados em formato de coluna.
  * Ganho de processamento e de espaco em disco.

# CSV vs Parquet
**CSV:**
* Nao existe estatisticas dos dados no arquivo.
* Eh preciso ler todo o arquivo para definicao dos data types.
* Nao permite compressao por coluna
* Ocupa mais espaco e mais processamento

### Convertendo arquivos CSV para Parquet com Spark

```
# Converte para o formato parquet
df.write.format("parquet")\
.mode("overwrite")\
.save("/FileStore/tables/processing/df-parquet-file-.parquet")
```

```
# lendo arquivo parquet
df_parquet = spark.read.format("parquet")\
.load("/FileStore/tables/processing/df-parquet-file-.parquet")
```

```
# conta a quantidade de linhas do dataframe
df_parquet.count()
```

### Scripts em SQL

* Criando uma tabela temporaria a partir do parquet (tbm pode ser feita com o csv)
* Utilizando codigo SQL para fazer Querys.

```
df_parquet.createOrReplaceTempView("tabela")
```

```
%sql
-- consulta a view criada
SELECT *
FROM tabela
WHERE Platform = 'Wii'
ORDER BY Global_Sales DESC
LIMIT 1
```
